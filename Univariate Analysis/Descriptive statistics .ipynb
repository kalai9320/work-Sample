# Descriptive statistics - Univariate Analysis

</font>

### Descriptive statistics are numerical measures that summarize and describe the main features of a dataset.They include measures such as measures of central tendency , measures of dispersion , measures of distribution shape. Descriptive statistics help in understanding the basic characteristics of the data without making any assumptions or predictions beyond the observed data.

```python
import pandas as pd
import numpy as np

# Creating a pandas Series with 30 data points
data = np.random.randint(0, 100, 30)
series = pd.Series(data)
# Descriptive statistics using pandas 
series.describe()
```

- This output provides descriptive statistics for each numerical column in the DataFrame, including count, mean, standard deviation, minimum, 25th percentile, median (50th percentile), 75th percentile, and maximum values.

# Measures of central tendency
</font>

- Measures of central tendency provide a summary of where the center of a dataset lies. They offer a single representative value that helps to understand the typical value or average within the data.

<font color='Blue'>

# Mean
    
</font>

### The mean, often referred to as the arithmetic average, is a measure of central tendency that represents the sum of all values in a dataset divided by the total number of values. It provides a single value that summarizes the entire dataset.

```python
import numpy as np
import pandas as pd

# Creating a pandas Series

series = pd.Series([85, 45, 40, 60, 50,  7, 
                    54, 49, 67, 36, 61, 89, 
                    41, 90, 19, 28, 71,34])

# Calculating the mean
mean = np.mean(series)

# Printing the mean
print('Mean = ', np.round(mean, 2))
```

<font color='Blue'> 

# Median
    
 </font>   

### The median is a measure of central tendency that represents the middle value of a dataset when it is ordered from smallest to largest. If there is an odd number of observations, the median is simply the middle value. However, if there is an even number of observations, the median is the average of the two middle values.

```python
import numpy as np
import pandas as pd

# Creating a pandas Series
series = pd.Series([85, 45, 40, 60, 50,  7, 
                    54, 49, 67, 36, 61, 89, 
                    41, 90, 19, 28, 71,34])

# Calculating the Median
Median = np.median(series)

# Printing the Median
print('Median = ', np.round(Median, 2))
```

# Mode
    
</font>

### In statistics, the mode refers to the value or values that occur most frequently in a dataset. The mode is useful for describing the most typical or common value in a dataset.

```python
from scipy import stats
import warnings
# Ignore warnings
warnings.filterwarnings("ignore")


# Generate some sample data
data = [10, 20, 30, 30, 40, 50, 50, 50, 35, 54, 50]

# Calculate the mode
mode_result = stats.mode(data)

# Print the mode
print("Mode:", mode_result.mode[0])
```

# Range
    
 </font>

### Range is difference between the maximum and minimum values in a dataset

```python
import numpy as np

# sample data
data = [10, 20, 30, 30, 40, 50, 50, 50, 35, 54, 50]

# Calculating the Range
Minimum = np.min(data)
Maximum = np.max(data)
Range = Maximum - Minimum

# Printing the Range
print("Range:", Range)
```

# variance
    
</font>

### variance measures  how much the values in a dataset vary from the mean. It's calculated by averaging the squared differences between each data point and the mean. 

```python
import numpy as np

# sample data
data = [10, 20, 30, 30, 40, 50, 50, 50, 35, 54, 50]

variance = np.var(data)
print('Variance: ',np.round((variance),2))
```

<font color='Blue'>

# Standard deviation
    
 </font>

### Standard deviation is the square root of the variance. It's often preferred over variance because it's in the same units as the original data, making it easier to interpret.

```python
import numpy as np

# sample data
data = [10, 20, 30, 30, 40, 50, 50, 50, 35, 54, 50]

Standard_deviation = np.std(data)
print('Variance: ',np.round((Standard_deviation),2))
```

# IQR-  Interquartile Range

 </font>

### This is the range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data when the dataset is sorted in ascending order. IQR is robust against outliers and provides a measure of dispersion that focuses on the middle 50% of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False)

# Calculate quartiles
quartiles = np.percentile(data, [25, 50, 75])

# Add text inside the box plot
plt.text(quartiles[0], 1.1, '25th Percentile: {:.2f}'.format(quartiles[0]), horizontalalignment='center')
plt.text(quartiles[1], 1.05, 'Median: {:.2f}'.format(quartiles[1]), horizontalalignment='center')
plt.text(quartiles[2], 1.1, '75th Percentile: {:.2f}'.format(quartiles[2]), horizontalalignment='center')

plt.title('Box Plot with Interquartile Range (IQR)')
plt.xlabel('Value')
plt.show()
```

# skewness
    
</font> 

### Skewness is a statistical measure that quantifies the asymmetry of the probability distribution of a dataset around its mean. In other words, it indicates whether the data is symmetrically distributed or skewed to one side.

## 1. Positive Skewness or Right Skewness
- In a positively skewed distribution, the tail on the right-hand side (higher values) is longer or fatter than the tail on the left-hand side (lower values). This indicates that the majority of the data points are concentrated on the left side of the distribution, with a few larger values pulling the mean to the right. The mean is typically greater than the median in a positively skewed distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data with positive skewness
data_positive_skew = np.random.gamma(shape=2, scale=1, size=1000)

# Plot histogram to visualize the distribution
plt.hist(data_positive_skew, bins=30, color='blue', alpha=0.7)
plt.title('Positive Skewness Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 2.Negative Skewness or Left Skewness
- In a negatively skewed distribution, the tail on the left-hand side (lower values) is longer or fatter than the tail on the right-hand side (higher values). This indicates that the majority of the data points are concentrated on the right side of the distribution, with a few smaller values pulling the mean to the left. The mean is typically less than the median in a negatively skewed distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data with negative skewness
data_negative_skew = -np.random.gamma(shape=2, scale=1, size=1000)

# Plot histogram to visualize the distribution
plt.hist(data_negative_skew, bins=30, color='red', alpha=0.7)
plt.title('Negative Skewness Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 3. Zero Skewness

-- A distribution is considered symmetric (or approximately symmetric) if it has zero skewness. This means that the distribution is balanced and the tails on both sides are of equal length.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data with zero skewness (normal distribution)
data_zero_skew = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram to visualize the distribution
plt.hist(data_zero_skew, bins=30, color='green', alpha=0.7)
plt.title('Zero Skewness Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# kurtosis

</font>  

#### kurtosis measures how much data is in the tails of a distribution and how sharp or flat the peak of the distribution is compared to the normal distribution.

 ##       1. Leptokurtic or Positive kurtosis: 
-  In a leptokurtic distribution, the tails of the distribution are heavier (or more peaked) than those of a normal distribution. This indicates that the data have more extreme values (outliers) than would be expected under a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data with positive kurtosis (leptokurtic distribution)
data_leptokurtic = np.random.laplace(loc=0, scale=1, size=1000)

# Plot histogram to visualize the distribution
plt.hist(data_leptokurtic, bins=30, color='purple', alpha=0.7)
plt.title('Leptokurtic Distribution (Positive Kurtosis)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 2. Platykurtic or negative kurtosis
- In a platykurtic distribution, the tails of the distribution are lighter (or less peaked) than those of a normal distribution. This indicates that the data have fewer extreme values (outliers) than would be expected under a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data with negative kurtosis (platykurtic distribution)
data_platykurtic = np.random.uniform(low=-2, high=2, size=1000)

# Plot histogram to visualize the distribution
plt.hist(data_platykurtic, bins=30, color='orange', alpha=0.7)
plt.title('Platykurtic Distribution (Negative Kurtosis)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# 3. Mesokurtic or zero kurtosis
- In a mesokurtic distribution, the tails of the distribution are similar in shape to a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data with zero kurtosis (mesokurtic distribution, normal distribution)
data_mesokurtic = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram to visualize the distribution
plt.hist(data_mesokurtic, bins=30, color='cyan', alpha=0.7)
plt.title('Mesokurtic Distribution (Zero Kurtosis)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Quartiles
    
</font>

- Quartiles divide a dataset into four equal parts, representing 25% of the data in each part.
- The first quartile (Q1) is the value below which 25% of the data falls.
- The second quartile (Q2), also known as the median, is the value below which 50% of the data falls.
- The third quartile (Q3) is the value below which 75% of the data falls.
- Quartiles are often used to summarize the spread of data and identify the central tendency.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = [10, 20, 30, 30, 40, 50, 50, 50, 35, 54, 50]

# Calculate quartiles
q1, q2, q3 = np.percentile(data, [25, 50, 75])

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False)

# Add text annotations for quartiles
plt.text(q1, 1.1, 'Q1', horizontalalignment='center')
plt.text(q2, 1.05, 'Q2'.format(q2), horizontalalignment='center')
plt.text(q3, 1.1, 'Q3'.format(q3), horizontalalignment='center')

# Add title and labels
plt.title('Box Plot with Quartiles')
plt.xlabel('Value')
plt.ylabel('Data')

# Remove y-axis ticks
plt.yticks([])

# Show plot
plt.show()
```

# Percentiles

</font>

- Percentiles divide a dataset into hundredths, representing 1% of the data in each part.
- The nth percentile is the value below which n% of the data falls.
- For example, the 25th percentile is the value below which 25% of the data falls.
- Percentiles offer even finer granularity for analyzing the distribution of data compared to quartiles and deciles.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = [10, 20, 30, 30, 40, 50, 50, 50, 35, 54, 50]

# Calculate percentiles
percentiles = np.percentile(data, [25, 50, 75])

# Plot histogram with percentiles
plt.hist(data, bins=10, color='skyblue', alpha=0.7)
plt.axvline(x=percentiles[0], color='red', linestyle='--', label='25th Percentile (Q1)')
plt.axvline(x=percentiles[1], color='green', linestyle='--', label='50th Percentile (Median)')
plt.axvline(x=percentiles[2], color='blue', linestyle='--', label='75th Percentile (Q3)')

# Add labels and legend
plt.title('Distribution with Percentiles')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Show plot
plt.show()
```

# Deciles

</font>

- Deciles divide a dataset into ten equal parts, representing 10% of the data in each part.
- The first decile (D1) is the value below which 10% of the data falls.
- The second decile (D2) is the value below which 20% of the data falls.
- Similarly, the ninth decile (D9) is the value below which 90% of the data falls.
- Deciles provide more detailed information about the distribution of data compared to quartiles.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate data from 10 to 100
data = np.arange(10, 101)

# Calculate percentiles
percentiles = np.percentile(data, range(10, 101, 10))

# Create distribution plot

sns.distplot(data, color='skyblue', kde=True, hist_kws=dict(edgecolor="black", linewidth=1))

# Add percentile lines
for percentile in percentiles:
    plt.axvline(x=percentile, color='red', linestyle='--')

# Add labels and legend
plt.title('Distribution Plot with Percentiles')
plt.xlabel('Value')
plt.ylabel('Density')

# Show plot
plt.legend([f'{i}th Percentile' for i in range(10, 101, 10)], loc='best')
plt.show()
```

#### These descriptive statistics are fundamental tools Univariant analysis used in data analysis to understand the distribution, variability, and relationships within a dataset, without making any assumptions beyond the data itself
